```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
```

```python
train_loss = pd.read_csv("../dvclive/plots/metrics/train/loss.tsv", sep="\t")
val_loss = pd.read_csv("../dvclive/plots/metrics/val/synth/loss.tsv", sep="\t")
```

```python
warnings.filterwarnings("ignore")

plt.style.use("seaborn")
plt.figure(figsize=(10, 5))
plt.plot(train_loss["step"], train_loss["loss"])
plt.plot(val_loss["step"], val_loss["loss"])
plt.legend(["train", "val"])
plt.title("Train Val Loss")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.show()
```
